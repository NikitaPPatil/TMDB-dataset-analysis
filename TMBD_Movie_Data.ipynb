```python
 def catigorize_col (df, col, labels):
    """
    catigorizes a certain column based on its quartiles
   
    Args:
        (df)     df   - dataframe we are proccesing
        (col)    str  - to be catigorized column's name 
        (labels) list - list of labels from min to max
    
    Returns:
        (df)     df   - dataframe with the categorized col
    """
    
    # setting the edges to cut the column accordingly
    edges = [df[col].describe()['min'],
             df[col].describe()['25%'],
             df[col].describe()['50%'],
             df[col].describe()['75%'],
             df[col].describe()['max']]
    
    df[col] = pd.cut(df[col], edges, labels = labels, duplicates='drop')
    return df
```

```python
# define labels for these edges
labels = ['not_popular', 'below_avg', 'average', 'popular']
catigorize_col(df, 'vote_average', labels)

df['vote_average'].unique()
```

```python
# dropping nans and confirming
df.dropna(inplace = True)
df.isnull().sum()
```

```python
# define labels for these edges
labels = ['low', 'average', 'high']
catigorize_col(df, 'profit', labels)

df['profit'].unique()
```

Q7: Which movies had the highest profit margins?

```python
# Calculating profit margin as (revenue - budget) / budget
df['profit_margin'] = (df['revenue'] - df['budget']) / df['budget']

# Sorting by profit margin and displaying the top 5 movies
top_profit_margin = df[['original_title', 'profit_margin']].sort_values(by='profit_margin', ascending=False).head(5)
top_profit_margin
```

Q8: What are the most profitable genres?

```python
# Grouping by genre and calculating the mean profit
genre_profit = df.groupby('genres')['profit'].value_counts(normalize=True).unstack().fillna(0)
genre_profit

# Visualizing the results
genre_profit.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title("Profitability by Genre", fontsize=14)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Proportion of Movies by Profit Category", fontsize=12)
plt.legend(title='Profit Category')
plt.grid(True)
plt.show()
```
